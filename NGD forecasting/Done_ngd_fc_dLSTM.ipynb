```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import acf
from statsmodels.graphics.tsaplots import plot_acf


# Read and preprocess the data
ngd_data = pd.read_csv('daily_NGD.csv', delimiter=',')
ngd_data['Date'] = pd.to_datetime(ngd_data['Date'])

type_ngd = 'IND' # Choose between 'NGD', 'LDZ', 'IND', 'GTP'
nr_lastdays = 30  # Number of days used for LSTM input
lags = 500
alpha = 0.05  # typically, a 95% confidence interval is used
period = 364  # default 364 days


variables_dict = {
    'NGD': ['HDD', 'HDD-1', 'Rus','War','Workday','Month','DE IND index'],
    'LDZ': ['HDD', 'HDD-1', 'Rus','War','Month','Workday','DE wind','DE solar','FR nuclear'],
    'IND': ['HDD', 'HDD-1', 'Rus','War','Month','Workday','DE wind','DE solar','DE IND index', 'FR nuclear','THE price'],
    'GTP': ['HDD', 'HDD-1', 'Rus','War','Month','DE wind','DE solar','FR nuclear','EU storage','EU LNG','THE price','DE IND index','UK NL production','African TAP']
}

if type_ngd not in variables_dict:
    raise ValueError("Invalid gas demand type selected!")

explanatory_vars = variables_dict[type_ngd]
ngd_data = ngd_data[['Date', type_ngd] + explanatory_vars]

# Compute ACF values
acf_values = acf(ngd_data[type_ngd], nlags=lags)

# Plot the ACF
plot_acf(ngd_data[type_ngd], lags=lags, alpha=alpha)
plt.show()

# Decompose the original data
result = seasonal_decompose(ngd_data[type_ngd], model='additive', period=period)
ngd_data['Trend'] = result.trend
ngd_data['Residual'] = result.resid
ngd_data['Seasonal'] = result.seasonal
ngd_data.dropna(inplace=True)
valid_index = ngd_data.dropna().index

ngd_data.to_csv('save_ngd_data.csv', index=True)  

# Create subplots
plt.figure(figsize=(14, 8))
plt.subplot(411)
plt.plot(ngd_data[type_ngd], label='Original Data')
plt.legend(loc='upper left')
plt.title('Original Data')
plt.subplot(412)
plt.plot(ngd_data['Trend'], label='Trend', color='orange')
plt.legend(loc='upper left')
plt.title('Trend')
plt.subplot(413)
plt.plot(result.seasonal, label='Seasonality', color='green')
plt.legend(loc='upper left')
plt.title('Seasonality')
plt.subplot(414)
plt.plot(ngd_data['Residual'], label='Residuals', color='red')
plt.legend(loc='upper left')
plt.title('Residuals')
plt.tight_layout()
plt.show()

def prepare_data(data, explanatory_vars, test_ratio):
    global nr_lastdays  # Access the global variable
    test_start_dt = round(len(data) * (1 - test_ratio)) + 1
    train = data.iloc[:test_start_dt]
    test = data.iloc[test_start_dt:]
    
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_train = scaler.fit_transform(train[explanatory_vars])
    scaled_test = scaler.transform(test[explanatory_vars])
    
    X_train, y_train = [], []
    for i in range(nr_lastdays, len(train)):
        X_train.append(scaled_train[i-nr_lastdays:i])
        y_train.append(scaled_train[i, 0])
    X_train, y_train = np.array(X_train), np.array(y_train)
    
    num_features = len(explanatory_vars)
    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], num_features))
    
    return train, test, X_train, y_train, scaler

# Adjusted for the Trend and Residual only as they are the target variables
train_trend, test_trend, X_train_trend, y_train_trend, scaler_trend = prepare_data(ngd_data, ['Trend'] + explanatory_vars[2:], 0.2)
train_residual, test_residual, X_train_residual, y_train_residual, scaler_residual = prepare_data(ngd_data, ['Residual'] + explanatory_vars[2:], 0.2)

# Build and compile the LSTM model
def build_model(input_shape):
    model = Sequential()
    model.add(LSTM(units=200, return_sequences=True, input_shape=input_shape))
    model.add(LSTM(units=200))
    model.add(Dropout(0.1))
    model.add(Dense(units=1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

model_trend = build_model((X_train_trend.shape[1], len(explanatory_vars) - 1))  # Excluding the target variable 'Trend'
model_residual = build_model((X_train_residual.shape[1], len(explanatory_vars) - 1))  # Excluding the target variable 'Residual'

# Train the models
model_trend.fit(X_train_trend, y_train_trend, epochs=50, batch_size=32)
model_residual.fit(X_train_residual, y_train_residual, epochs=50, batch_size=32)

def inverse_transform_target(scaler, predictions, target_idx=0):
    # Create dummy array with the same shape as training data
    dummy = np.zeros((len(predictions), len(scaler.scale_)))
    dummy[:, target_idx] = predictions.squeeze()
    # Inverse transform using scaler
    inv_transformed = scaler.inverse_transform(dummy)
    return inv_transformed[:, target_idx]

# Predict the Trend and Residuals
def predict(model, train, test, scaler, X_train, explanatory_vars, nr_lastdays):
    predicted_train = model.predict(X_train)
    dataset_total = pd.concat((train[explanatory_vars], test[explanatory_vars]), axis=0)
    inputs = dataset_total.values[len(dataset_total) - len(test) - nr_lastdays:]
    inputs = scaler.transform(inputs)
    
    X_test = []
    for i in range(nr_lastdays, nr_lastdays + len(test)):
        X_test.append(inputs[i-nr_lastdays:i])
    X_test = np.array(X_test)
    num_features = len(explanatory_vars)
    X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], num_features))
    
    predicted_demand = model.predict(X_test)
    predicted_train = inverse_transform_target(scaler, predicted_train)
    predicted_demand = inverse_transform_target(scaler, predicted_demand)

    return predicted_train, predicted_demand

predicted_trend_train, predicted_trend_demand = predict(model_trend, train_trend, test_trend, scaler_trend, X_train_trend, ['Trend'] + explanatory_vars[2:], nr_lastdays)
predicted_residual_train, predicted_residual_demand = predict(model_residual, train_residual, test_residual, scaler_residual, X_train_residual, ['Residual'] + explanatory_vars[2:], nr_lastdays)

# Add forecasted trend and residual to original seasonality
forecasted_train = predicted_trend_train + predicted_residual_train + result.seasonal[:len(predicted_trend_train)]
forecasted_demand = predicted_trend_demand + predicted_residual_demand + result.seasonal[-len(predicted_trend_demand):]

# Truncate forecasted_train to match the length of train_trend['Date'][60:]
forecasted_train = forecasted_train[:len(train_trend['Date'][nr_lastdays:])]

colors = sns.color_palette("colorblind")

# Visualize original trend vs. forecasted trend
plt.figure(figsize=(16, 6))
plt.scatter(ngd_data['Date'], ngd_data['Trend'], color=colors[1], label='Original Trend')
plt.plot(train_trend['Date'][nr_lastdays:], predicted_trend_train, color=colors[2], label='Forecasted Train Trend')
plt.plot(test_trend['Date'], predicted_trend_demand, color=colors[3], label='Forecasted Test Trend')
plt.title('Original Trend vs Forecasted Trend')
plt.legend()
plt.show()

# Visualize original residual vs. forecasted residual
plt.figure(figsize=(16, 6))
plt.scatter(ngd_data['Date'], ngd_data['Residual'], color=colors[1], label='Original Residual')
plt.plot(train_residual['Date'][nr_lastdays:], predicted_residual_train, color=colors[2], label='Forecasted Train Residual')
plt.plot(test_residual['Date'], predicted_residual_demand, color=colors[3], label='Forecasted Test Residual')
plt.title('Original Residual vs Forecasted Residual')
plt.legend()
plt.show()

# Visualize results
plt.figure(figsize=(16, 6))
plt.scatter(ngd_data['Date'], ngd_data[type_ngd], color=colors[1], label='Original')
plt.plot(train_trend['Date'][nr_lastdays:], forecasted_train, color=colors[2], label='Train Forecast')
plt.plot(test_trend['Date'], forecasted_demand, color=colors[3], label='Test Forecast')
plt.legend()
plt.show()

# Evaluation
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

train_mape = mean_absolute_percentage_error(train_trend[type_ngd].iloc[-len(forecasted_train):], forecasted_train)
test_mape = mean_absolute_percentage_error(test_trend[type_ngd], forecasted_demand)
train_rmse = np.sqrt(mean_squared_error(train_trend[type_ngd].iloc[nr_lastdays:], forecasted_train))
test_rmse = np.sqrt(mean_squared_error(test_trend[type_ngd], forecasted_demand))
train_r2 = r2_score(train_trend[type_ngd].iloc[nr_lastdays:], forecasted_train)
test_r2 = r2_score(test_trend[type_ngd], forecasted_demand)
train_mae = mean_absolute_error(train_trend[type_ngd].iloc[nr_lastdays:], forecasted_train)
test_mae = mean_absolute_error(test_trend[type_ngd], forecasted_demand)

# Print the results
print(f"Evaluation {type_ngd} based on variables: {explanatory_vars}")
print(f"LSTM nr_lastdays: {nr_lastdays}")
print(f"Period: {period}")
print("Training MAPE: {:.4f}%".format(train_mape))
print("Testing MAPE: {:.4f}%".format(test_mape))
print("Training RMSE: {:.4f}".format(train_rmse))
print("Testing RMSE: {:.4f}".format(test_rmse))
print("Training R2 score: {:.4f}".format(train_r2))
print("Testing R2 score: {:.4f}".format(test_r2))
print("Training MAE: {:.4f}".format(train_mae))
print("Testing MAE: {:.4f}".format(test_mae))
```

```python
# Code to forecast the next days with variables forecased using Prophet

from prophet import Prophet
import holidays

# Define the number of days you want to predict into the future
future_days = 60

# Define variables to forecast with Prophet
variables_to_forecast = explanatory_vars

# Dictionary to hold future forecasts
future_forecasts = {}

# Forecast future values for each variable using Prophet
for var in variables_to_forecast:
    if var not in ['Rus', 'War', 'Workday']:
        prophet_df = ngd_data[['Date', var]].rename(columns={'Date': 'ds', var: 'y'})
        m = Prophet(daily_seasonality=True, yearly_seasonality=True, weekly_seasonality=True)
        m.fit(prophet_df)
        future = m.make_future_dataframe(periods=future_days)
        forecast = m.predict(future)
        future_forecasts[var] = forecast['yhat'][-future_days:].values

# Handle constant variables like 'Rus' and 'War'
for var in ['Rus', 'War']:
    last_value = ngd_data[var].iloc[-1]
    future_forecasts[var] = np.full((future_days,), last_value)

# Calculate 'Workday' for future dates
if 'Workday' in variables_to_forecast:
    future_dates = pd.date_range(start=ngd_data['Date'].iloc[-1] + pd.Timedelta(days=1), periods=future_days, freq='D')
    years = future_dates.year.unique()
    german_holidays_obj = holidays.Germany(years=years.tolist())
    german_holidays = list(german_holidays_obj.keys())
    future_forecasts['Workday'] = (~future_dates.weekday.isin([5, 6]) & ~future_dates.isin(german_holidays)).astype(int)

# Convert to DataFrame
future_forecasts_df = pd.DataFrame(future_forecasts)

# Scale the future forecasts
scaler_exog = MinMaxScaler(feature_range=(0, 1))
scaler_exog.fit(ngd_data[explanatory_vars[2:]])
scaled_future_vars = scaler_exog.transform(future_forecasts_df[explanatory_vars[2:]])

# Prepare the future input for LSTM
X_future = []
for i in range(future_days):
    if i < nr_lastdays:
        hist_slice = ngd_data[explanatory_vars[2:]].iloc[-(nr_lastdays-i):].values
        future_slice = scaled_future_vars[:i+1]

        # Reshape slices to ensure they are 2D
        hist_slice = hist_slice.reshape(-1, len(explanatory_vars) - 2)
        future_slice = future_slice.reshape(-1, len(explanatory_vars) - 2)

        # Combine slices
        if i == 0:
            combined_slice = hist_slice
        else:
            combined_slice = np.vstack((hist_slice, future_slice))
    else:
        combined_slice = scaled_future_vars[(i-nr_lastdays+1):(i+1)]
        combined_slice = combined_slice.reshape(-1, len(explanatory_vars) - 2)

    X_future.append(combined_slice)

# Convert X_future to an array
X_future = np.array(X_future)

# Ensure that X_future is a 3D array
if X_future.ndim != 3:
    raise ValueError(f"X_future should be a 3D array, got {X_future.ndim} dimensions")

# Reshape X_future for LSTM input
X_future = np.reshape(X_future, (X_future.shape[0], X_future.shape[1], len(explanatory_vars) - 2))

# Predict future trend and residuals using the LSTM models
predicted_future_trend = model_trend.predict(X_future)
predicted_future_residual = model_residual.predict(X_future)

# Inverse transform the predictions
predicted_future_trend = inverse_transform_target(scaler_trend, predicted_future_trend, target_idx=0)
predicted_future_residual = inverse_transform_target(scaler_residual, predicted_future_residual, target_idx=0)

# Combine the predicted trend and residuals with the seasonal component
seasonal_component = ngd_data['Seasonal'].iloc[-1]
final_future_forecast = predicted_future_trend + predicted_future_residual + seasonal_component

# Create a DataFrame for the forecast
forecast_dates = pd.date_range(start=ngd_data['Date'].iloc[-1] + pd.Timedelta(days=1), periods=future_days, freq='D')
forecast_df = pd.DataFrame({
    'Date': forecast_dates,
    'Forecasted_NGD': final_future_forecast.flatten()
})

# Visualize the forecast
plt.figure(figsize=(16, 6))
plt.plot(forecast_df['Date'], forecast_df['Forecasted_NGD'], color='blue', label='Forecasted Gas Demand')
plt.title('Forecasted Gas Demand')
plt.legend()
plt.show()

# Save the forecast to a CSV file
forecast_df.to_csv('forecasted_gas_demand.csv', index=False)

# Print the forecast
print(forecast_df)

```
